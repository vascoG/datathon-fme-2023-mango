```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the outfit data from the CSV file
outfit_data = pd.read_csv('../dataset/outfit_data.csv')

product_data = pd.read_csv('../dataset/product_data.csv')

# Loop through each row in the outfit data
for outfit in outfit_data.itertuples():
    # Get the filename of the image

    product_id = outfit.cod_modelo_color
    product = product_data.loc[product_data['cod_modelo_color'] == product_id]    

    if product.des_age.values[0] != 'Kids':
        continue

    image_filename = product['des_filename'].values[0]
    image_filename = "../../" + image_filename


# Check if the image file exists
    if os.path.isfile(image_filename):
        # Load the image and display it
        image = plt.imread(image_filename)
        plt.imshow(image)
        plt.show()
    else:
        print(f"Image file {image_filename} does not exist.")

    
```
