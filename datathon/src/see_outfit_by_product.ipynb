```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the outfit data from the CSV file
outfit_data = pd.read_csv('../dataset/outfit_data.csv')

product_data = pd.read_csv('../dataset/product_data_clean.csv')

# Change here the product you want
new_product_data = product_data[product_data.des_sex == 'Male']

# Merge the outfit data with the product data
for product in new_product_data.itertuples():
    print("NEW PRODUCT\n")
    outfit_data_tmp = outfit_data[outfit_data.cod_modelo_color == product.cod_modelo_color]

    if len(outfit_data_tmp) > 0:
        for outfit in outfit_data_tmp.itertuples():
            print("NEW OUTFIT\n")
            outfit_tmp = outfit_data[outfit_data.cod_outfit == outfit.cod_outfit]
            for tmp_product in outfit_tmp.itertuples():
                product_id = tmp_product.cod_modelo_color

                new_product = product_data[product_data['cod_modelo_color'] == product_id]  

                if new_product.empty:
                    continue

                image_filename = new_product['des_filename']
                image_filename = "../../" + image_filename
                image_filename = image_filename.values[0]

                # Check if the image file exists
                if os.path.isfile(image_filename):
                    # Load the image and display it
                    image = plt.imread(image_filename)
                    plt.imshow(image)
                    plt.show()
                else:
                    print(f"Image file {image_filename} does not exist.")

```
